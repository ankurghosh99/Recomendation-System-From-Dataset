```python
import json
```

```python
def load_data(filename):
    with open (filename,"r") as f:
        return json.load(f)

data = load_data("data_for_pro3.json")
```

```python
data
```

```python
user_id = 1

def page_u_may_like(user_id,data):
    user_pages={}
    for user in data["users"]:
        user_pages[user['id']] = set(user['liked_pages'])
    if user_id not in user_pages:
        return []
    user_liked_pages = user_pages[user_id]
    page_suggetion = {}
    for other_user,liked_pages in user_pages.items():
        if other_user != user_id:
            shared_pages = user_liked_pages.intersection(liked_pages)
        for liked_page in liked_pages:
            if liked_page not in user_liked_pages:
                page_suggetion[liked_page] = page_suggetion.get(liked_page,0)+ len(shared_pages)
                
    sorted_pages = sorted(page_suggetion.items(),key = lambda x:x[1],reverse = True)
    return[(suggested_page,score) for (suggested_page,score) in sorted_pages]
            

recc = page_u_may_like(user_id,data)
print(recc)
```
